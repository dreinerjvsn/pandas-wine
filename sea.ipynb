```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.get_dataset_names()
iris = sns.load_dataset("iris")
```

### Explorando histogramas e gráficos distributivos no DataSet Iris

```python
iris
```

#### Como o tamanho da pétala varia nas diferentes espécies de <i>iris<i>

```python
sns.histplot(data=iris,x=iris['petal_length'],kde=True)
plt.title("Tamanho de pétalas (cm)")
```

```python
sns.histplot(data=iris,x=iris['petal_length'],hue=iris['species'],kde=True)
plt.title("Tamanho de pétalas (cm) por espécies")
```

```python
#Kernel Density Plot - Histograma suavizado
sns.kdeplot(data=iris['petal_length'],fill=True)
```

 Percebe-se que a linha evidenciada no histograma anterior se mantem no gráfico KDE. Há uma relação de pétalas de, aproximadamente, 1cm e de 5cm em um maior número de espécimes

#### Vamos agora visualizar isso em um _JointPlot_ que correlacionará, não apenas, o tamanho das pétalas já visto, mas também a largura das sépalas:
- Esse Joitplot pode ser resumido em um 2D KDEplot

```python
sns.jointplot(x=iris['petal_length'],y=iris['sepal_width'],kind="kde")
```

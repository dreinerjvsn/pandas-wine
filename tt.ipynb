```python
import pandas as pd
```

```python
reviews = pd.read_csv("winemag-data-130k-v2.csv",index_col=0)
reviews
```

```python
bargain = (reviews.points / reviews.price).idxmax()
out = reviews.loc[bargain,"title"]
out
```

```python
def contaPalavra(palavra):
    count = 0
    for txt in reviews['description']:
        if palavra in txt:
            count += 1
    return count;
    
trop = contaPalavra("tropical")
frt = contaPalavra("fruity")
print(trop,frt)

descriptor_counts = pd.Series({"tropical": trop,"fruity":frt})
descriptor_counts
```

#### Fazendo um Series que represente uma nova pontuação

```python
def atribuirPontuacao(linha):    
    if linha["country"] == "Canada":
        return 3;
    elif linha["points"] >= 95:
        return 3;
    elif linha['points'] >= 85 & linha['points'] < 95:
        return 2;
    else:
        return 1;

rating = reviews.apply(atribuirPontuacao, axis=1)
rating
```

```python
sumarizacao = reviews.groupby('price').points.max()
sumarizacao.idxmax()
reviews[reviews['points']==sumarizacao.idxmax()]
```

#### Caso queiramos pegar o melhor vinho de cada país, independente da província ou demais índices

```python
melhor_pais = reviews.groupby("country").apply(lambda reviews: reviews.loc[reviews['points'].idxmax()])
melhor_pais
```

```python
taster_twitter = reviews.groupby("taster_twitter_handle").taster_twitter_handle.value_counts()
taster_twitter
```

```python
best_rating_per_price = reviews.groupby('price').points.max()
best_rating_per_price.sort_values()
best_rating_per_price
```

```python
precos_tipo = reviews.groupby('variety').price.agg(['min','max'])
precos_tipo
```

#### Plotando charts sobre atrasos de vôos para linhas aéreas americanas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
voos = pd.read_csv("flight_delays.csv", index_col="Month")
#voos = voos.fillna(0)
#voos_validos = voos[voos[0:] > 0.0]
#voos_validos = voos.dropna(subset=voos["US"])
voos
```

```python
sns.lineplot(data=voos,x="Month",y="AA")
plt.figure(figsize=(12,6))
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title("Atraso médio dos voôs da Spirit Airlines por mês")
bar = sns.barplot(data=voos,x=voos.index,y="NK")
bar.bar_label(bar.containers[0],fontsize=8)
plt.show()
```

#### Heatmaps

```python
plt.figure(figsize=(14,7))
sns.heatmap(data=voos,annot=True)
plt.title("Atrasos em Linhas Aéreas")
plt.show()
```

A partir do HeatMap acima, notamos que a linha aérea NK foi a que teve o maior atraso em um mês durante o ano inteiro, com um atraso de, aproximadamente, 36.

Além disso, podemos notar nesse tipo de gráfico, quais os meses em que cada linha aérea obteve seu maior atraso

- O que pode ser comprovado na série abaixo:

```python
voos.iloc[0:][0:].idxmax()
```

#### Qual linha aérea teve uma maior quantidade de atrasos por mês?

```python
linhas = {}
for linha in voos.columns:
    linhas[linha] = 0
atrasados = pd.Series(data=linhas)
atrasados

for mon in voos.index:
    atrasado = voos.loc[mon].idxmax()
    atrasados.loc[atrasado] += 1
atrasados
```

#### Conclui-se que a companhia mais atrasada é a "Spirit Airlines - NK"

```python
#Set-Up
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

sns.set_theme(style="white")
warnings.filterwarnings("ignore")

tips = sns.load_dataset("tips")
```

```python
print(tips.dtypes)
print(tips.describe())
tips.head(5)
```

```python
sns.scatterplot(x=tips['tip'],y=tips['total_bill'])
plt.title("Gorgetas pela conta paga")
plt.xlabel("Gorgetas")
plt.ylabel("Conta Refeição")
plt.show()
```

```python
sns.regplot(x=tips['tip'],y=tips['total_bill'])
```

```python
sns.swarmplot(x=tips['sex'],y=tips['tip'],palette='dark:#5A9_r')
```
